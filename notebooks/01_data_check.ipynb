```python
# âœ… Imports
import sys, os
sys.path.append(os.path.abspath("../src"))

import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

from data_loader import load_and_prepare_data
from model import LSTMModel
from train import train_model

# âœ… Define evaluation function locally (safe version)
def evaluate_model(model, X, y):
    model.eval()
    with torch.no_grad():
        outputs = model(X)
        
        if outputs.dim() > 1:
            preds = outputs.squeeze(-1)
        else:
            preds = outputs

        preds = preds.detach().cpu().numpy()
        truth = y.detach().cpu().numpy()

        print("RMSE:", np.sqrt(mean_squared_error(truth, preds)))
        print("MAE:", mean_absolute_error(truth, preds))
        print("RÂ² Score:", r2_score(truth, preds))

        plt.figure(figsize=(10, 5))
        plt.plot(truth[:200], label="Actual")
        plt.plot(preds[:200], label="Predicted")
        plt.legend()
        plt.title("Bandwidth Prediction for Indian Locations")
        plt.grid(True)
        plt.show()

# âœ… Load and prepare data
X, y, _ = load_and_prepare_data("../data/simulated_indian_5g.csv", sequence_length=10)

# âœ… Create model
model = LSTMModel(input_size=7)

# âœ… Train model
train_model(model, X, y, epochs=50)

# âœ… Evaluate
evaluate_model(model, X, y)
```
